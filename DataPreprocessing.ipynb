```python
# Handle outliers using IQR instead of manually handling outliers
def handle_outliers(df_sleep, feature):
    Q1 = df_sleep[feature].quantile(0.25)
    Q3 = df_sleep[feature].quantile(0.75)
    IQR =Q3-Q1
    minimum = Q1=(1.5*IQR)
    maximum = Q3+(1.5*IQR)
    df_sleep[feature] = df_sleep[feature].clip(minimum, maximum)
    return df_sleep

numerical_columns = [col for col in numerical_features if df_sleep[col].nunique()>10]
for col in numerical_columns:
    df_sleep = handle_outliers(df_sleep, col)

df_sleep.head()
```

```python
numerical_features = ['Total Sleep Hours', 'Sleep Quality', 'Exercise (mins/day)',
                     'Caffeine Intake (mg)', 'Screen Time Before Bed (mins)',
                     'Work Hours (hrs/day)', 'Mood Score', 'Stress Level',
                     'Productivity Score']
categorical_features = ['Gender']
time_features = ['Sleep Start Time', 'Sleep End Time'] 

# Scaling numerical features
numerical_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())  # or RobustScaler() if outliers are a concern
])

# One-Hot Encoding for categorical features
categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(sparse_output=False, handle_unknown='ignore')) 
])

# Cyclical encoding for time features
'''Sine and cosine functions are inherently cyclical. They produce wave-like patterns that repeat over a given period. 
We can use this property to represent time in a way that preserves its cyclical nature.'''
for feature in time_features:
    df_sleep[feature + '_sin'] = np.sin(2 * np.pi * df_sleep[feature] / 24)
    df_sleep[feature + '_cos'] = np.cos(2 * np.pi * df_sleep[feature] / 24)

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_features),
        ('cat', categorical_transformer, categorical_features),
    ],
    remainder='passthrough'  # Include other features (cyclical encoded time)
)
features_to_preprocess = numerical_features + categorical_features + time_features + [f'{feature}_sin' for feature in time_features] + [f'{feature}_cos' for feature in time_features]
sleep_df_preprocessed = preprocessor.fit_transform(df_sleep[features_to_preprocess])

#print(df_sleep.head())
```
