```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering, KMeans
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score

# Load the dataset
df = pd.read_csv('preprocessed_data.csv')

# Select features for unsupervised learning
features = ['Total Sleep Hours', 'Sleep Quality', 'Exercise (mins/day)',
            'Caffeine Intake (mg)', 'Screen Time Before Bed (mins)',
            'Mood Score', 'Stress Level', 'Productivity Score']
X = df[features]

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 1. Hierarchical Clustering
hierarchical_cluster = AgglomerativeClustering(n_clusters=3)
cluster_labels_hierarchical = hierarchical_cluster.fit_predict(X_scaled)

# Visualize the hierarchical clustering results (using total sleep hours and sleep quality as an example)
plt.figure(figsize=(10, 6))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=cluster_labels_hierarchical, cmap='viridis')
plt.xlabel('Total Sleep Hours (Scaled)')
plt.ylabel('Sleep Quality (Scaled)')
plt.title('Hierarchical Clustering Results')
plt.show()

# 2. K-means Clustering
kmeans = KMeans(n_clusters=3, random_state=42)
cluster_labels_kmeans = kmeans.fit_predict(X_scaled)

# Evaluate K-means (Silhouette Score)
silhouette = silhouette_score(X_scaled, cluster_labels_kmeans)
print(f'K-means Silhouette Score: {silhouette}')

# 3. PCA Dimensionality Reduction
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Visualize the PCA results
plt.figure(figsize=(10, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=cluster_labels_kmeans, cmap='viridis')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.title('PCA Visualization of Clusters')
plt.show()
```
