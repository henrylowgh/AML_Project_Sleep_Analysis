```python
#Supervised Learning

ss = StandardScaler()
ss.fit(X_dev_ohe)
dev_scaled = ss.transform(X_dev_ohe)
test_scaled = ss.transform(X_test_ohe)

dev_preprocessed = np.hstack([np.ones((dev_scaled.shape[0], 1)), dev_scaled])
test_preprocessed = np.hstack([np.ones((test_scaled.shape[0], 1)), test_scaled])

dev_preprocessed
```

```python
class LinearRegression():
    '''
    Linear regression model with L2-regularization (i.e. ridge regression).

    Attributes
    ----------
    alpha: regularization parameter
    w: (n x 1) weight vector
    '''
    def __init__(self, alpha=0):
        self.alpha = alpha
        self.w = None

    def train(self, X, y):
        '''Trains model using ridge regression closed-form solution.
        Parameters:
        X : (m x n) feature matrix
        y: (m x 1) label vector
        '''
        ### Your code here
        x_t = np.transpose(X)
        x_t_x = np.dot(x_t, X)
        a_i = self.alpha * np.identity(x_t_x.shape[0])
        part1 = np.linalg.inv(x_t_x + a_i)
        part2 = np.dot(part1, x_t)
        self.w = np.dot(part2, y)

    def predict(self, X):
        '''Predicts on X using trained model.
        Parameters:
        X : (m x n) feature matrix
        Returns:
        y_pred: (m x 1) prediction vector
        '''
        ### Your code here
        return np.dot(X, self.w)
        
```

```python
def get_report(y_pred, y_test):
    """
    Report the first 3 and last 3 predictions . on X_test,
    along with the actual labels in y_test.

    Returns:
        A dataframe with 6 rows comparing predictions and actuals.
    """
    preds = np.concatenate([y_pred[:3], y_pred[-3:]])
    actuals = np.concatenate([y_test[:3], y_test[-3:]])
    df_compare = pd.DataFrame({'Prediction': preds, 'Actual': actuals})
    df_compare['Position'] = [1, 2, 3, len(y_pred) - 2, len(y_pred) - 1, len(y_pred)]
    df_compare = df_compare.set_index('Position')
    return df_compare
```

```python

lr_mood = LinearRegression()
lr_mood.train(dev_preprocessed, y_dev_mood)
y_dev_pred_mood = lr_mood.predict(dev_preprocessed)
y_test_pred_mood = lr_mood.predict(test_preprocessed)

lr_stress = LinearRegression()
lr_stress.train(dev_preprocessed, y_dev_stress)
y_dev_pred_stress = lr_stress.predict(dev_preprocessed)
y_test_pred_stress = lr_stress.predict(test_preprocessed)

lr_productivity = LinearRegression()
lr_productivity.train(dev_preprocessed, y_dev_productivity)
y_dev_pred_productivity = lr_productivity.predict(dev_preprocessed)
y_test_pred_productivity = lr_productivity.predict(test_preprocessed)


r2_train_mood = r2_score(y_dev_mood, y_dev_pred_mood)
r2_test_mood = r2_score(y_test_mood, y_test_pred_mood)
display(get_report(lr_mood.predict(test_preprocessed), y_test_mood))
print(f"Mood r2 for training set: {r2_train_mood}; r2 for test set: {r2_test_mood}")

r2_train_stress = r2_score(y_dev_stress, y_dev_pred_stress)
r2_test_stress = r2_score(y_test_stress, y_test_pred_stress)
display(get_report(lr_stress.predict(test_preprocessed), y_test_stress))
print(f"Stress r2 for training set: {r2_train_stress}; r2 for test set: {r2_test_stress}")

r2_train_productivity = r2_score(y_dev_productivity, y_dev_pred_productivity)
r2_test_productivity = r2_score(y_test_productivity, y_test_pred_productivity)
display(get_report(lr_productivity.predict(test_preprocessed), y_test_productivity))
print(f"Productivity r2 for training set: {r2_train_productivity}; r2 for test set: {r2_test_productivity}")
```

```python
class LogisticRegression():
    '''
    Logistic regression model with L2 regularization.

    Attributes
    ----------
    alpha: regularization parameter
    t: number of epochs to run gradient descent
    eta: learning rate for gradient descent
    w: (n x 1) weight vector
    '''

    def __init__(self, alpha=0, t=100, eta=1e-3):
        self.alpha = alpha
        self.t = t
        self.eta = eta
        self.w = None

    def train(self, X, y):
        '''Trains logistic regression model using gradient descent
        (sets w to its optimal value).

        Parameters
        ----------
        X : (m x n) feature matrix
        y: (m x 1) label vector

        Returns
        -------
        losses: (t x 1) vector of losses at each epoch of gradient descent
        '''

        loss = list()
        self.w = np.zeros((X.shape[1],1))
        for i in range(self.t):
            self.w = self.w - (self.eta * self.calculate_gradient(X, y))
            loss.append(self.calculate_loss(X, y))
        return loss

    def predict(self, X):
        '''Predicts on X using trained model. Make sure to threshold
        the predicted probability to return a 0 or 1 prediction.

        Parameters
        ----------
        X : (m x n) feature matrix

        Returns
        -------
        y_pred: (m x 1) 0/1 prediction vector
        '''
        y_pred = self.calculate_sigmoid(X.dot(self.w))
        y_pred[y_pred >= 0.5] = 1
        y_pred[y_pred < 0.5] = 0
        return y_pred

    def calculate_loss(self, X, y):
        '''Calculates the logistic regression loss using X, y, w,
        and alpha. Useful as a helper function for train().

        Parameters
        ----------
        X : (m x n) feature matrix
        y: (m x 1) label vector

        Returns
        -------
        loss: (scalar) logistic regression loss
        '''
        return -y.T.dot(np.log(self.calculate_sigmoid(X.dot(self.w)))) - (1-y).T.dot(np.log(1-self.calculate_sigmoid(X.dot(self.w)))) + self.alpha*np.linalg.norm(self.w, ord=2)**2

    def calculate_gradient(self, X, y):
        '''Calculates the gradient of the logistic regression loss
        using X, y, w, and alpha. Useful as a helper function
        for train().

        Parameters
        ----------
        X : (m x n) feature matrix
        y: (m x 1) label vector

        Returns
        -------
        gradient: (n x 1) gradient vector for logistic regression loss
        '''
        return X.T.dot(self.calculate_sigmoid( X.dot(self.w)) - y) + 2*self.alpha*self.w


    def calculate_sigmoid(self, x):
        '''Calculates the sigmoid function on each element in vector x.
        Useful as a helper function for predict(), calculate_loss(),
        and calculate_gradient().

        Parameters
        ----------
        x: (m x 1) vector

        Returns
        -------
        sigmoid_x: (m x 1) vector of sigmoid on each element in x
        '''
        return (1)/(1 + np.exp(-x.astype('float')))
```
